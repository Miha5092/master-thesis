### Some results

```python
v5 = np.mean(np.load('results/predictions/vitae/random_unet_random_5_predictions.npz')['errors'])
v10 = np.mean(np.load('results/predictions/vitae/random_unet_random_10_predictions.npz')['errors'])
v15 = np.mean(np.load('results/predictions/vitae/random_unet_random_15_predictions.npz')['errors'])
v20 = np.mean(np.load('results/predictions/vitae/random_unet_random_20_predictions.npz')['errors'])
v25 = np.mean(np.load('results/predictions/vitae/random_unet_random_25_predictions.npz')['errors'])
v30 = np.mean(np.load('results/predictions/vitae/random_unet_random_30_predictions.npz')['errors'])

v5_cnn = np.mean(np.load('results/predictions/vcnn/random_classic_random_5_predictions.npz')['errors'])
v10_cnn = np.mean(np.load('results/predictions/vcnn/random_classic_random_10_predictions.npz')['errors'])
v15_cnn = np.mean(np.load('results/predictions/vcnn/random_classic_random_15_predictions.npz')['errors'])
v20_cnn = np.mean(np.load('results/predictions/vcnn/random_classic_random_20_predictions.npz')['errors'])
v25_cnn = np.mean(np.load('results/predictions/vcnn/random_classic_random_25_predictions.npz')['errors'])
v30_cnn = np.mean(np.load('results/predictions/vcnn/random_classic_random_30_predictions.npz')['errors'])

v5_unet = np.mean(np.load('results/predictions/vcnn/random_tiny_random_5_predictions.npz')['errors'])
v10_unet = np.mean(np.load('results/predictions/vcnn/random_tiny_random_10_predictions.npz')['errors'])
v15_unet = np.mean(np.load('results/predictions/vcnn/random_tiny_random_15_predictions.npz')['errors'])
v20_unet = np.mean(np.load('results/predictions/vcnn/random_tiny_random_20_predictions.npz')['errors'])
v25_unet = np.mean(np.load('results/predictions/vcnn/random_tiny_random_25_predictions.npz')['errors'])
v30_unet = np.mean(np.load('results/predictions/vcnn/random_tiny_random_30_predictions.npz')['errors'])
```

```python
plt.figure(figsize=(8, 6))

plt.plot([5, 10, 15, 20, 25, 30], [v5, v10, v15, v20, v25, v30], marker="o", label="Vitae UNet")
plt.plot([5, 10, 15, 20, 25, 30], [v5_cnn, v10_cnn, v15_cnn, v20_cnn, v25_cnn, v30_cnn], marker="o", label="VCNN")
plt.plot([5, 10, 15, 20, 25, 30], [v5_unet, v10_unet, v15_unet, v20_unet, v25_unet, v30_unet], marker="o", label="V-UNet")
plt.xlabel("Number of sensors")
plt.ylabel("Average Relative error")

plt.grid()
plt.legend()
plt.show()
```
